### Pixel Sum and Comparison

```python
#Find maximum in each image (which is within the particle's area) and sum values of all pixels in a window (dx,dy) around it
lista=[Array1,Array2,Array3,Array4,Array5]
dy=10
dx=10
fig, axs=plt.subplots(1,5,figsize=(22,9))
plt.subplots_adjust(wspace=0.3)
for k,array in enumerate(lista):
    m=np.max(array)
    s=np.where(array==m)
    ym=s[0][0]
    xm=s[1][0]
    c=array[ym-dy:ym+dy,xm-dx:xm+dx]
    #c=array[507:530,670:695]
    #plt.figure()
    print('\033[1mPicture\033[0m',k+1,'Maximum:',np.max(c),'- Minimum:',np.min(c),' - Coordinates Max(y,x)', ym,',', xm,
          ' - Sum of pixels:',np.sum(c), ', (Tot #pixels',c.size,')')
    axs[k].imshow(c, cmap='gray')
    
    #Convert axis values of the plots back to the original 
    h1 = np.arange(0,15,5)
    h2 = np.arange(ym-dy,ym+dy,2)
    h3 = np.arange(xm-dx,xm+dx,2)
    axs[k].set_yticks(h1)
    axs[k].set_yticklabels(h2.tolist())
    axs[k].set_xticks(h1)
    axs[k].set_xticklabels(h3.tolist())
    
```

```python
#Pick one particle and fit the two cross-sections to a Lorentzian or Gaussian
A=Array5
m=np.max(A)
s=np.where(A==m)
dx=12
dy=12
ym=s[0][0]
xm=s[1][0]
c=A[ym-dy:ym+dy,xm-dx:xm+dx]
h=np.where(c==np.max(c))
Y=h[0][0]
X=h[1][1]
print(h,Y,X)

ascissa = np.arange(0,24,1)   #For the fitting
ascissa1 =np.arange(0,24,0.1) #For the plotting

def Lorentz(w,b,a,Wc,k):
    return b+a*(k/2)**2/((w-Wc)**2+(k/2)**2)
#def Gauss(r,a,b,f):
 #   return 20+a*np.exp(-np.power(r-b,2)/(2*np.power(f,2)))

param =[10,150,12,1.5] 
fitto1, cov1 = curve_fit(Lorentz,ascissa,c[Y,:],param)
fitto2, cov2 = curve_fit(Lorentz,ascissa,c[:,X],param)
print('b:',fitto1[0],'a:',fitto1[1],'Center:',fitto1[2],'k:',fitto1[3])
print('b:',fitto2[0],'a:',fitto2[1],'Center:',fitto2[2],'k:',fitto2[3])

fig, (ax0,ax1, ax2)=plt.subplots(1,3,figsize=(20,5))
plt.subplots_adjust(wspace=0.3)
ax0.imshow(c)
ax0.hlines(Y,0,23, color='c', linestyle='--')
ax0.vlines(X,0,23, color='c', linestyle='--')

ax1.plot(c[Y,:],'o')
ax1.plot(ascissa1, Lorentz(ascissa1,fitto1[0],fitto1[1],fitto1[2],fitto1[3]))
ax1.set_ylim(bottom=10, top=m+20)
ax1.set_title('Horizontal', fontsize=12, style='italic')

ax2.plot(c[:,X],'o')
ax2.plot(ascissa1, Lorentz(ascissa1,fitto2[0],fitto2[1],fitto2[2],fitto1[3]))
ax2.set_ylim(bottom=10, top=m+20)
ax2.set_title('Vertical', fontsize=12, style='italic');
```
